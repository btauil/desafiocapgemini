##  Questão 1 <a class="anchor" id="Questão 1"></a>

Escreva um algoritmo que mostre na tela uma escada de tamanho **n** utilizando o caractere * e espaços. A base e altura da escada devem ser iguais ao valor de **n**. A última linha não deve conter nenhum espaço.

#### Exemplo:

**Entrada:**

`
n = 6
`

**Saída:**

```
     *
    **
   ***
  ****
 *****
******
```

### Resposta

A solução para esse problema é criar um loop capaz de adicionar espaços em branco e * em linhas. Cada corrida do loop ele adiciona um * e retira um espaço em branco.

````python
def escada():
    # Input do tamanho n, checa para valores válidos
    e = ""
    try:
        n = int(input("Digite o tamanho da escada: "))
    except ValueError:
        print("Não é um valor aceito")
        return
    
    # for loop responsável pela escada
    for i in range(1,n+1):
        offset = n - i #espaço em branco
        e += " " * offset + "*" * i #inserindo os espaços e os *
        print(e)
        e = "" #resetando o string
        
# Driver
escada()
````

##  Questão 2 <a id="Questão 2"></a>

Débora se inscreveu em uma rede social para se manter em contato com seus amigos. A página de cadastro exigia o preenchimento dos campos de nome e senha, porém a senha precisa ser forte. O site considera uma senha forte quando ela satisfaz os seguintes critérios:

- Possui no mínimo 6 caracteres.
- Contém no mínimo 1 digito.
- Contém no mínimo 1 letra em minúsculo.
- Contém no mínimo 1 letra em maiúsculo.
- Contém no mínimo 1 caractere especial. Os caracteres especiais são: **!@#$%^&*()-+**

Débora digitou uma string aleatória no campo de senha, porém ela não tem certeza se é uma senha forte. Para ajudar Débora, construa um algoritmo que informe qual é o número mínimo de caracteres que devem ser adicionados para uma string qualquer ser considerada segura.

#### Exemplo:

**Entrada:**

`
Ya3
`

**Saída:**

`
3
`

**Explicação:**

Ela pode tornar a senha segura adicionando 3 caracteres, por exemplo, &ab, transformando a senha em Ya3&ab. 2 caracteres não são suficientes visto que a senha precisa ter um tamanho mínimo de 6 caracteres.

### Resposta

O algoritmo para resolver esse problema deve checar para cada um dos requisitos exigidos pela rede social. Faz isso através de uma contagem de pontos, onde cada requisito não cumprido vale por 1 ponto. O algoritmo separa os pontos entre caracteres que faltam para o tamanho mínimo e os outros requisitos e escolhe o valor com mais pontos no final. Essa divisão evita o algoritmo a errar a pontuação em situações como:

Entrada:
`
@!
`

Saída:
`
7
`

````python
import re

def verificar_senha():
    
    # Pedir um input do usuario
    senha = input("Coloque a senha: ")

    # Contagem para checagem dos requerimentos, T para o tamnho e R para os outros
    pontosT = 0 
    pontosR = 0
    
    # Ifs para checar o string de senha para cada requerimento
    # Tamanho do string
    if len(senha) < 6:
        pontosT += 6 - len(senha) # adiciona no contador de pontos
    
    # Digitos
    if not re.search(r"\d", senha):
        pontosR += 1
        
    # Maiúsculas
    if not re.search(r"[A-Z]", senha):
        pontosR += 1
    
    # Minúsculas
    if not re.search(r"[a-z]", senha):
        pontosR += 1
    
    # Caracteres Especiais
    if not re.search("[!@#$%^&*()+-]", senha):
        pontosR += 1

    # Checar para situações onde 1 mesmo caractere cumpre um requerimento e alcança o tamanho mínimo 
    print(max(pontosT, pontosR))

# Driver
verificar_senha()
````

##  Questão 3 <a id="Questão 3"></a>

Duas palavras podem ser consideradas anagramas de si mesmas se as letras de uma palavra podem ser realocadas para formar a outra palavra. Dada uma string qualquer, desenvolva um algoritmo que encontre o número de pares de substrings que são anagramas.


#### Exemplo 1:

**Entrada:**

`
ovo
`

**Saída:**

`
2
`

**Explicação:**

A lista de todos os anagramas pares são: [o, o], [ov, vo] que estão nas posições [[0], [2]], [[0, 1], [1, 2]] respectivamente. 

#### Exemplo 2:

**Entrada:**

`
ifailuhkqq
`

**Saída:**

`
3
`

**Explicação:**

A lista de todos os anagramas pares são: [i, i], [q, q] e [ifa, fai] que estão nas posições [[0], [3]], [[8],  [9]] e [[0, 1, 2], [1, 2, 3]].

### Resposta

O objetivo desse problema é encontrar o total de anagramas substring dentro de um string específico, podendo ser resolvido com um mapa. O algoritmo vai fazer uma contágem das frequências de cada substring e depois vai calcular a quantidade de pares anagramas.

````python
s = "ifailuhkqq"
mp = dict()

# Loop pelo tamanho do string
for i in range(len(s)):
    sb = "" # Define o substring
    
    # Loop para formar um dicionário com cada substring do string original e quantas vezes ele aparece
    for j in range(i, len(s)):
        sb = "".join(sorted(sb + s[j])) # 
        mp[sb] = mp.get(sb, 0)
        mp[sb] += 1

# Contagem de anagramas
anas = 0

# Loop passa por todos os itens do dicionário e faz a contagem de anagramas
for k, v in mp.items():
    anas += (v*(v-1))//2
print(anas)
````
